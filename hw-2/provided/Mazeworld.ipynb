## A* Search and Multi-robot Coordination
In this section we will discuss our implementation of A* Search in the context of the multi-robot planning problem. First, we will test our implementation on the test maze2 which is a small 4x4 maze with only one robot. For this test we will also be using our null hueristic making this A* search simply a Uniform Cost Search.

```python
search_problem_2 = MazeworldProblem.MazeworldProblem(maze2, (2,2))
result_2 = astar_search.astar_search(search_problem_2, search_problem_2.null_heuristic)
```

Now lets look at the path:

```python
search_problem_2.animate_path(result_2.path)
```

We can now look at a more complicated problem in maze3, still using uniform cost search:

```python
search_problem_3 = MazeworldProblem.MazeworldProblem(maze3, (1, 4, 1, 3, 1, 2))
result_3 = astar_search.astar_search(search_problem_3, search_problem_3.null_heuristic)
```

Looking more concisely at the state path as opposed to the animation we have:

```python
print(result_3.path)
print(len(result_3.path))
```

Now using a different heuristic, say manhattan distance, we can find the following path:

```python
#-- Manhattan Distance
result_3man = astar_search.astar_search(search_problem_3, search_problem_3.manhattan_heuristic)
print(result_3man.path)
print(len(result_3man.path))
```

We can see that using a better beuristic for this problem led to a shorter path to the goal. The uniform cost search took 27 moves, where as the manhattan heuristic only took 19. 

Now lets look at some larger and more interesting mazes. First, lets look at maze that just consists of a long corridor in which the robots will be required to change position:


```python
corridor = Maze.Maze(os.path.join(os.getcwd(), 'provided/corridor_maze.maz'))
print(corridor)

#-- Let's search the corridor and have our goal state be the re-order of the robot positions.
search_corr = MazeworldProblem.MazeworldProblem(corridor, goal_locations=(37,3,36,3,35,3))
```

We can explore this search space using both the manhattan and uniform heuristic and compare their performances. 

```python
result_corr_man = astar_search.astar_search(search_corr, search_corr.manhattan_heuristic)
result_corr_null = astar_search.astar_search(search_corr, search_corr.null_heuristic)
print('Manhattan Path Length:', len(result_corr_man.path))
print('Uniform Cost Search Path Length:', len(result_corr_null.path))
```

Clearly, the manhattan heuristic performs far better than the uniform cost search in terms of finding a shortest path.

Now we will set up all the search problems for the runtime test based on the size of the maze with proportional walls

```python
search_problem_rand_1 = MazeworldProblem.MazeworldProblem(random_10x10, goal_locations= random_10x10.get_random_goal_state())
search_problem_rand_2= MazeworldProblem.MazeworldProblem(random_20x20, goal_locations= random_20x20.get_random_goal_state())
search_problem_rand_3= MazeworldProblem.MazeworldProblem(random_30x30, goal_locations= random_30x30.get_random_goal_state())
search_problem_rand_4 = MazeworldProblem.MazeworldProblem(random_40x40, goal_locations= random_40x40.get_random_goal_state())
search_problem_rand_5= MazeworldProblem.MazeworldProblem(random_80x80, goal_locations= random_80x80.get_random_goal_state())
search_problem_rand_6 = MazeworldProblem.MazeworldProblem(random_120x120, goal_locations= random_120x120.get_random_goal_state())
search_problems = [search_problem_rand_1, search_problem_rand_2, search_problem_rand_3,
                  search_problem_rand_4, search_problem_rand_5, search_problem_rand_6]
```

```python
avg_times_man = []
for search_problem in search_problems:
    times = []
    for _ in range(1):
        start_time = time.time()
        result = astar_search.astar_search(search_problem, search_problem.manhattan_heuristic)
        times.append(time.time() - start_time)
    avg_times_man.append(np.average(times))
```

```python
df_man = pandas.DataFrame([[10,20,30,40,80,120], avg_times_man])
df_man = df_man.transpose()

ax = sns.regplot(df_man[0],df_man[1], color='r', order=2, truncate=True)
ax.set_xlabel('Maze Size (nxn)')
ax.set_ylabel('Run Time (sec)')
ax.set_title('Runtime Analysis on Maze Size')
```
